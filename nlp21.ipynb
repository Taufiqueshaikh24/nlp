# 2. Calculate the avg word length, avg sentence length (in words) and lexical diversity for "Moby Dick" by Herman Melville using gutenberg corpus.


```python
from nltk.corpus import gutenberg
from nltk import word_tokenize, sent_tokenize
from nltk.probability import FreqDist

moby_dick = gutenberg.words('melville-moby_dick.txt')
moby_dick_sentences = gutenberg.sents('melville-moby_dick.txt')

word_lengths = [len(word) for word in moby_dick if word.isalpha()]
avg_word_length = sum(word_lengths) / len(word_lengths)

sentence_lengths = [len(sentence) for sentence in moby_dick_sentences]
avg_sentence_length = sum(sentence_lengths) / len(sentence_lengths)

lexical_diversity = len(set(moby_dick)) / len(moby_dick)

avg_word_length, avg_sentence_length, lexical_diversity
```

# 7. Using the inaugural corpus plot a frequency distribution of the word 'freedom' over the years.


```python
import matplotlib.pyplot as plt
from nltk import FreqDist

years = inaugural.fileids()

freedom_counts = []
for year in years:
    words = inaugural.words(year)
    fdist = FreqDist(words)
    freedom_counts.append(fdist['freedom'])

plt.figure(figsize=(20,10))
plt.plot(years, freedom_counts)
plt.xlabel('Year')
plt.ylabel('Frequency of "freedom"')
plt.title('Frequency of the word "freedom" in US Inaugural Addresses')
plt.xticks(rotation=45)
plt.show()
```
